```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("concrete_data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
target = df['Strength']
predictors = df.drop('Strength', axis = 1)
```

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors.head()
```

```python
n_cols = predictors_norm.shape[1]
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def regression_model():
  model = Sequential()
  model.add(Dense(10, activation='relu', input_shape = (n_cols,)))
  model.add(Dense(10,activation="relu"))
  model.add(Dense(1))

  model.compile(optimizer = 'adam', loss = 'mean_squared_error')
  return model
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, train_size=0.3)
```

```python
model= regression_model()
```

```python
model.fit(X_train, y_train, epochs = 100,verbose=0)
```

```python
results = model.evaluate(X_test, y_test)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
```

```python
mse_list = []
for i in range(50):
  X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, train_size=0.3)
  model= regression_model()
  model.fit(X_train, y_train, epochs = 100,verbose=0)
  results = model.evaluate(X_test, y_test)
  y_pred = model.predict(X_test)
  mse = mean_squared_error(y_test, y_pred)
  mse_list.append(mse)
```

```python
mse_list
```

```python
mean_mse = np.mean(mse_list)
std_mse = np.std(mse_list)
print(f"The mean of the mean squared errors is {mean_mse:.2f}.")
print(f"The standard deviation of the mean squared errors is {std_mse:.2f}.")
```
